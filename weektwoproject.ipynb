e. Get the national ID and name of all students who have cleared their fees (fee_cleared)


```python
%%sql SELECT national_id,financial_details.stud_name
FROM personal_details INNER JOIN financial_details
ON personal_details.stud_ID = financial_details.stud_ID
WHERE fee_cleared = 'TRUE';
```

g. Get the count of students who share a current home county i.e., Say Nairobi, get the number of students whos current_home_county is Nairobi, and so on for all available counties (home_county_count)

```python
%%sql SELECT current_home_county, COUNT(current_home_county)
FROM school_details
GROUP BY current_home_county;
```

h. Get the count of Male and/or Female students from each secondary_school_county (secondary_school_count). The table should contain a column for male student count and female student count for each county.

```python
%%sql SELECT current_home_county,
COUNT(case when gender='Male' then 1 end) AS male_student_count,
COUNT(case when gender='Female' then 1 end) AS female_student_count
FROM school_details INNER JOIN personal_details
ON school_details.stud_ID=personal_details.stud_ID
GROUP BY current_home_county;
```

i. Get the percentage of students who set their next_of_kin as Mother vs those that set it as Father1. (kin_percentage)

```python
%%sql
 SELECT (COUNT(case when next_of_kin_relation='Mother' then 1 end)/COUNT(next_of_kin_relation))*100 AS mother_percentage,
 (COUNT(case when next_of_kin_relation='Father' then 1 end)/COUNT(next_of_kin_relation))*100 AS father_percentage
 FROM contact_details;
```
